```python
import pandas as pd
import numpy as np
from glob import glob
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
from PIL import Image
import easyocr
```

```python
images_path = glob('images/*')
```

```python
reader = easyocr.Reader(['en'],gpu = True) 
```

```python
imgs = images_path[2] # To test on the spot
```

```python
result = reader.readtext(imgs)
```

```python
pd.DataFrame(result, columns=['bbox','text','conf'])
```

```python
def showcase_results(img, result):
    image = Image.open(img)

    fig, axes = plt.subplots(1, 2, figsize=(10, 10))

    axes[0].imshow(image)
    axes[0].set_title('Original Image')
    axes[0].axis('off')
    
    axes[1].imshow(image)
    for (bbox, text, prob) in result:
    
        (top_left, top_right, bottom_right, bottom_left) = bbox
        poly = plt.Polygon([top_left, top_right, bottom_right, bottom_left], edgecolor='red', fill=None)
        axes[1].add_patch(poly)
        plt.text(top_left[0], top_left[1] - 10, text, fontsize=12, color='red', bbox=dict(facecolor='white', alpha=0.5))
        
    axes[1].set_title('Image with OCR Results')
    axes[1].axis('off')
    plt.axis('off')
    plt.show()
```

```python
showcase_results(imgs,result)
```

```python
############################################################################################################
```

```python
final_results=[]

for imgs in tqdm(images_path[:5]):
    result = reader.readtext(imgs)
    showcase_results(imgs,result)
    img_id = imgs.split('\\')[-1].split('.')[0]
    img_df = pd.DataFrame(result, columns=['bbox','text','conf'])
    img_df['img_id'] = img_id
    final_results.append(img_df)
prediction_df = pd.concat(final_results)
```
